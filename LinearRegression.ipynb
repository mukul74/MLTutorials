```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
dia_X, dia_Y = datasets.load_diabetes(return_X_y=True)
```

```python
print(dia_X.shape)
print(dia_Y.shape)
```

```python
dia_x_one = dia_X[:,2:3]
print(dia_x_one.shape)
plt.scatter(dia_x_one, dia_Y)
```

```python
dia_x_one_small = dia_x_one[:100]
dia_Y_small = dia_Y[:100]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dia_x_one_small, dia_Y_small, test_size=0.25, random_state=42)
```

```python
print(f"Shape of Train data and Test data : {X_train.shape, X_test.shape}")
```

```python
l_reg = linear_model.LinearRegression()
```

```python
l_reg.fit(X_train, y_train)
```

```python
print(f"Coeff and intercept for the linear regression model : {l_reg.coef_.item(), l_reg.intercept_}")
```

```python
y_pred = l_reg.predict(X_test)
```

```python
plt.scatter(X_train, y_train,color='g') 
plt.plot(X_test, y_pred,color='k') 
```
